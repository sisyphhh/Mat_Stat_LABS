```python
def otrezok(massiv_xi):
    otrezzzki=[]
    a0=massiv_xi[0]
    am=massiv_xi[-1]
    #dlinna=len(massiv_xi)
    d=am-a0
    d_del_m=d/m
    otrezzzki.append(a0)
    i=1
    while i<m:
        ai=d_del_m + otrezzzki[i-1]
        otrezzzki.append(round(ai,5))
        i+=1
    otrezzzki.append(am)
    print('Otrezzki: ',otrezzzki,'\n')
    return otrezzzki
```

```python
def gruppirovka(lin2,a__x):
    a__n=[]
    ran,i=0,0
    j=1#right border
    while i<len(lin2):
        if lin2[i]<=a__x[j]:
            #i+=1
            ran+=1
            if i != len(lin2)-1:
                i+=1
                continue
        a__n.append(ran)
        ran=0
        if lin2[i]>a__x[j]:
            j+=1
            continue
        if i == len(lin2)-1:
            break    
    return a__n
```
