```python
print("variant 11")
import random
import numpy as np
from matplotlib import pylab as plt
from matplotlib import rcParams
%matplotlib inline
```

```python
v=11
n=5 + v%17
p=0.1 + 0.01*v
liam=0.7 + 0.07*v
liam = float('{:.5f}'.format(liam))
p = float('{:.5f}'.format(p))#func showing limited numb after dot
print('v=',v,' ','n=',n,' ','p=',p,' ','liambda=',liam)
```

```python
#Zadanie 1 math
#line 1,2 (xi,ni)

lin2_bin=[]#Xi
lin1_bin=[]#Ni
lin3_bin=[]#Wi

lin1_bin.clear()
lin2_bin.clear()
lin3_bin.clear()

for i in range(150):
    #ran=random.randint(0, 16)
    ran=np.random.binomial(n, p)
    #lin1.append(i)
    lin2_bin.append(ran)
   
print('Our massive: ',lin2_bin)
lin2_bin.sort()
print('Sorted massive',lin2_bin)
    
#delete copies
i=0
while 1:
    ran2 = lin2_bin.count(lin2_bin[i])/150
    ran2 = float('{:.5f}'.format(ran2))
    lin3_bin.append(ran2)
    #print(lin3,i)
    
    if lin2_bin.count(lin2_bin[i])>1 :
        kk=1
        while lin2_bin.count(lin2_bin[i])>1:
            kk+=1
            lin2_bin.pop(i)
        lin1_bin.append(kk)
        if i<len(lin2_bin):
            i+=1
            #kk+=1
    else:
      #kk+=1
        lin1_bin.append(1)
        i+=1
      
    #if i>=len(lin2) or kk>14:
    if i>=len(lin2_bin):
        break

print('Xi=',lin2_bin)
print('Ni=',lin1_bin)
print('Wi=',lin3_bin)
print(sum(lin3_bin))
             
```

```python
#rewrite like function
def calculll(lina1,lina2,lina3):

    x_=0
    for i in range(len(lina2)):
        x_+=lina2[i]*lina3[i]
    x_ = float('{:.5f}'.format(x_))
    print('Выборочное среднее :',x_)

    #for future calculations
    Nu1=0 
    Nu2=0 
    Nu3=0
    Nu4=0

    D_=0  #D_=Nu2-Nu1**2
    for i in range(len(lina2)):
        Nu1+=lina2[i]*lina3[i]
        Nu2+=(lina2[i]**2)*lina3[i]
        Nu3+=(lina2[i]**3)*lina3[i]
        Nu4+=(lina2[i]**4)*lina3[i]

    D_=Nu2 - (Nu1)**2
    D_ = float('{:.5f}'.format(D_))
    print('Выборочная дисперсия :',D_)

    #Центральный момент к-ого порядка:
    Nu1_0=0
    Nu2_0=D_
    Nu3_0=Nu3-3*Nu2*Nu1+2*(Nu1)**3  #=Nu3-3*Nu2*Nu+2*(Nu1)**3
    Nu4_0=Nu4-4*Nu3*Nu1+6*Nu2*(Nu1)**2-3*(Nu1)**4  #=Nu4-4*Nu3*Nu1+6*Nu2*(Nu1)**2-3*(Nu1)**4

    O_=D_**(0.5)
    O_ = float('{:.5f}'.format(O_)) 
    print('Выборочное среднее квадратическое отклонение :',O_)

    Y1_=(Nu3_0)/(O_**3)
    Y1_ = float('{:.5f}'.format(Y1_))
    print('Выборочный Коэффициент асимметрии :',Y1_)

    Y2_=(Nu4_0)/(O_**4)-3
    Y2_ = float('{:.5f}'.format(Y2_))
    print('Выборочный Коэффициент эксцесса :',Y2_)

    Moda0=0
    Moda_mass=[lina1[0]]#1 elemet of Ni
    Moda_index=[0]#index of first element
    trigger='zero'
    #trigger2=-1
    for i in range(len(lina1)):#1 4 3 12 12 9 12 55, len == 8 last index ==7

        if i==len(lina1)-1:
            break

        elif Moda_mass[Moda0]>lina1[i+1]:#12 12 9 12
            if i<len(lina1)-2 :
                if Moda_mass[Moda0]==lina1[i+2]:#случай если в послед-ти максимумов есть меньшее знач
                    #Moda_mass.clear()
                    Moda_index.clear()#~del Moda_index[:]
                    Moda_index.append(i)#запомнили индекс макс из ряда 
                    #если список не закончен ждем последующего возможного максимума
                    #trigger2=i
                    trigger='none'
                else:
                    continue
            continue

        elif Moda_mass[Moda0]<lina1[i+1]:
            trigger='zero'
            Moda_mass.clear()#del Moda_mass[:]
            Moda_index.clear()#del Moda_index[:]
            Moda_mass.append(lina1[i+1])#частота числа из выборки
            Moda_index.append(i+1)

        elif Moda_mass[Moda0]==lina1[i+1]:
            #Moda_mass.append(lin1_bin[i+1])#частота числа из выборки
            Moda_index.append(i+1)
            #Moda0+=1
    if trigger == 'none':
        Moda0='none'
    else:
        if len(Moda_index)==1:
            Moda0=lina2[Moda_index[0]]#Moda_index хранит индекс макс элемента/ов
        else:
            Moda0=(0.5)*(lina2[Moda_index[0]]+lina2[Moda_index[-1]])
            
    print('Выборочная мода :',Moda0)

    median_mass=[]
    box=0
    for i in range(len(lina2)):#emperic func
        box+=lina3[i]
        median_mass.append(box)

    Median=0
    i=0
    while i<len(lina2):
        if median_mass[i]==0.5 and i!=len(lina2)-1:
            Median=0.5*(lina2[i]+lina2[i+1])
            break
        elif i!=0 and median_mass[i-1]<0.5 and median_mass[i]>0.5:
            Median=lina2[i]
            break
        else:
            i+=1  
    print('Выборочное медиана :',Median)   
```

```python
def imperik_func(x,y):#masss2-Xi(lin2_bin) masss3-Wi(lin3_bin)
    xdel=[]
    ydel=[]
    #x=masss2
    #y=masss2
    boxdel=0
    plt.title('Empirical distribution function')
    plt.ylabel('sum(Wi)')
    plt.xlabel('Xi')
    plt.grid(True)
    for i in range(len(x)):
        if i == len(x)-1:
            xdel.append(x[i])
            xdel.append(x[i]+5)
            boxdel+=y[i]
            ydel.append(boxdel)
            ydel.append(boxdel)
        else:
            xdel.append(x[i])
            xdel.append(x[i+1])
            boxdel+=y[i]
            ydel.append(boxdel)
            ydel.append(boxdel)

        plt.plot(xdel,ydel,color='black')
        xdel.clear()
        ydel.clear()
    plt.show()
```

```python
def bino(masss):
    y=[]
    for i in range(len(masss)):
        n=16
        p=0.21
        q=0.79
        answww=1
        pp=1
        qq=1
        
        indx=masss[i]
        #print(indx)
        
        if indx==0:
            for jo in range(n-indx):
                answww*=(1-p)
                #answww*=q
            answww = float('{:.5f}'.format(answww))
            y.append(answww)
            
        elif indx <= n//2:
            indx2=indx
            #print(indx2)
            for j1 in range(indx):
                answww*=n/(indx2)#change indx to j1+1
                #print(n,'/',indx2)
                n-=1
                indx2-=1#delete
                pp*=p
            n=16#vosstanavili
            for j2 in range(n-indx): 
                qq*=(1-p)
                #qq*=q
            #print(n-indx,'   ',answww,'   ',pp,'   ',qq)
            answww*=pp
            answww*=qq
            #print(answww)
            answww = float('{:.5f}'.format(answww))
            y.append(answww)
            
        elif indx > n//2:
            indx=n-indx
            indx2=indx
            for j1 in range(indx):
                answww*=n/(indx2)#change indx to j1+1
                n-=1
                indx2-=1#delete
                qq*=(1-p)
            for j2 in range(n-indx): 
                pp*=p                
            answww*=pp
            answww*=qq
            answww = float('{:.5f}'.format(answww))
            y.append(answww)
        #print(y)
    return y
```

```python
def geomi(masss):
    y=[]
    n=16
    p=0.21
    for i in range(len(masss)):
        answww=1
        pp=1
        qq=1
        
        indx=masss[i]
        
        if indx==0:
            answww=p
            y.append(answww)         
        else :    
            for j1 in range(indx):
                answww*=(1-p)        
            answww*=p
            answww = float('{:.5f}'.format(answww))
            y.append(answww)      
    return y
```

```python
def pus(masss):
    y=[]
    e=2.718
    liambda=1.47
    
    for i in range(len(masss)):
       
        answww=1
        
        indx=masss[i]
        
        if indx==0:
            answww=(1/e)**liambda
            answww = float('{:.5f}'.format(answww))
            y.append(answww)         
        else :    
            for j1 in range(indx):
                answww*=liambda/(j1+1) 
            answww*=(1/e)**liambda
            answww = float('{:.5f}'.format(answww))
            y.append(answww) 
    
    return y   
```
